```python
#daily returns
df_tesla['daily_return'] = (df_tesla['Adj Close'] / df_tesla['Adj Close'].shift(1)) - 1

#calculate weekly returns
df_tesla['weekly_return'] = (df_tesla['Adj Close'].resample('W-Fri').last() / df_tesla['Adj Close'].resample('W-Fri').first()) - 1

#calculate monthly returns
df_tesla['monthly_return'] = (df_tesla['Adj Close'].resample('M').last() / df_tesla['Adj Close'].resample('M').first()) - 1

plt.figure(figsize=(12, 8))
#plot daily returns
plt.subplot(3, 1, 1)
plt.plot(df_tesla.index, df_tesla['daily_return'], marker='o')
plt.title('Daily Returns')

#plot weekly returns
plt.subplot(3, 1, 2)
plt.plot(df_tesla.index, df_tesla['weekly_return'], marker='o')
plt.title('Weekly Returns')

#plot monthly returns
plt.subplot(3, 1, 3)
plt.plot(df_tesla.index, df_tesla['monthly_return'], marker='o')
plt.title('Monthly Returns')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

df_google2['Date'] = pd.to_datetime(df_google2['Date'])
df_google2 = df_google2.sort_values(by='Date')
train_percentage = 0.8
split_index = int(len(df_google2) * train_percentage)
train_data = df_google2.iloc[:split_index]
test_data = df_google2.iloc[split_index:]

X_train = np.array(train_data['Date']).astype(np.int64).reshape(-1, 1)
y_train = train_data['Adj Close']

X_test = np.array(test_data['Date']).astype(np.int64).reshape(-1, 1)
y_test = test_data['Adj Close']

#create a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

#calculate mse of this model
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error on Testing Set:", mse)
slope = model.coef_[0]
print("Slope:", slope)
#plot the model
plt.figure(figsize=(12, 6))
plt.scatter(df_google2['Date'], df_google2['Adj Close'], label='Adj Close', color='red', s=10)
plt.plot(test_data['Date'], predictions, label='Linear Regression Model', color='blue')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```

I am going to create a linear model for each stock using its historical data. This model is relatively simple and likely will not perform well due to the volatility of the stocks but it will give a rough estimate of the direction which the stock will head in. The other model will be an ETS model which has a higher degree of complexity which will be able to capture more of the intricate aspects of the data. It will likely be more accurate as it takes error, trend, and seasonality into account. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error

df_google2['Date'] = pd.to_datetime(df_google2['Date'])
df_google2 = df_google2.sort_values(by='Date')

#training and testing sets
train_percentage = 0.8
split_index = int(len(df_google2) * train_percentage)
train_data = df_google2.iloc[:split_index]
test_data = df_google2.iloc[split_index:]

# ETS model
model = ExponentialSmoothing(train_data['Adj Close'], trend='add', seasonal='multiplicative', seasonal_periods=12)
fit_model = model.fit()

# predictions
predictions = fit_model.forecast(len(test_data))

# MSE
mse = mean_squared_error(test_data['Adj Close'], predictions)
print("Mean Squared Error on Testing Set:", mse)

# plot the results
plt.figure(figsize=(12, 6))
plt.scatter(df_google2['Date'], df_google2['Adj Close'], label='Actual Adj Close', color='red', s=10)
plt.plot(test_data['Date'], predictions, label='ETS Model', color='blue')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.title('Google Linear Model')
plt.legend()
plt.show()
```
